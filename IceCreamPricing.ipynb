## The Wacky World of Pricing Psychology

__Price Anchoring:__ As the saying goes, the best way to sell a \\$2,000 watch is to put it right next to a \\$10,000 watch. But why? The culprit is a common cognitive bias called anchoring. Anchoring refers to the tendency to heavily rely on the first piece of information offered when making decisions….Placing premium products and services near standard options may help create a clearer sense of value for potential customers, who will view the less expensive options as a bargain in comparison. (https://www.helpscout.com/blog/pricing-strategies/ Gregory Ciotti, May 15th, 2016)

What does this mean for Food Lion’s ice cream isle?  Large sections of that freezer are very similar brands, selling for approximately the same amount. Faced with many similar choices, many customers will experience “analysis paralysis”, and simply unconsciously decline to consider any ice cream.  By adding very expensive, although admittedly tasty. $4.95 pints of premium ice cream to the mix the grocer might well not expect to sell very many of those pints.  But once a customer considers the premium brand, and immediately rejects it as being too expensive, they have engaged in the analytic process of considering which ice cream to purchase.  Once begun, that analysis is more likely to continue until a purchase is made.


__Stickiness of Sale Price:__ One of the difficulties of offering items “on-sale”, especially if that item has an established market price, is that the sale price can erode that established price if it is too memorable.  A pint of ice cream that everyone “expects” to sell for \\$5 could easily be harmed it customers began to question that price because of a sale selling it for \\$3.  But a sale of 3 for \\$9, doesn’t create the same memory - rather it was a discounted price as a reward for buying “in bulk”.


__Comparative Pricing__ can Hurt Sales Selling by price comparison alone can end up hurting sales as there is a natural assumption that price is directly related to quality.  Think about it: When selling in the US, the Yugo was drastically cheaper than competing cars - yet they never had a significant volume in sales.

So, in my saga of my ice cream purchase, I find it fascinating that I received an identical coupon for Talenti - another premium ice cream typically priced exactly the same as Ben and Jerry’s.
